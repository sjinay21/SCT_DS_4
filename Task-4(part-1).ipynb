### Task-4 -- Part 1:-Data Loading & Preprocessing

This part includes loading the US Accidents dataset, checking for missing values, and extracting time-based features for further analysis.

```python
# Data handling
import pandas as pd

# Load dataset
df = pd.read_csv('US_Accidents_March23.csv')

# Display basic information
df.info()
df.head()
```

```python
# Check missing values
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])

# Drop rows with missing Start_Time and Weather_Condition
df = df.dropna(subset=['Start_Time', 'Weather_Condition'])
```

```python
# Convert to datetime
df['Start_Time'] = pd.to_datetime(df['Start_Time'], errors='coerce')
df['Hour'] = df['Start_Time'].dt.hour

df[['Start_Time', 'Hour']].head()
```
